```python
import numpy as np 
import pandas as pd
data = pd.read_excel("Data analyst Data.xlsx")
df = data
df.head()
```

```python
df.rename(columns = {'First Name' : 'Name'}, inplace  = True)
df.rename(columns = {'How did you come to know about this event?' : 'Source'}, inplace = True)
df.rename(columns ={'Specify in "Others" (how did you come to know about this event)' : 'Other Sources'}, inplace = True)
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
s = df['Name'].nunique()
print("The number of unique Students in the data set is : ", s )
```

```python
gpa = np.mean(df['CGPA'])
print("The average CGPA of Students is :", gpa)
```

```python
import matplotlib.pyplot as plt
Year = df['Year of Graduation'].value_counts()
print("The distribution of Students across different years are as follows :\n", Year)
#we can also plot it as a graph
plt.bar(Year.index,Year.values, color = "brown")
plt.xlabel("Year of Joining")
plt.ylabel("Number of Students Joined")
plt.title("Joining Year of Students")
plt.xticks(Year.index)

plt.show()
```

```python
p = df['Experience with python (Months)'].value_counts()
print("The distribution of students experince in python is as follows :\n", p)
#we can visualize it by
plt.bar(p.index, p.values, color = "green")
plt.xticks(p.index)
plt.title("Experince of Students in Python")
plt.xlabel("Years of Experience in Python")
plt.ylabel("Number of Students")
plt.show()

```

```python
import seaborn as sns

plt.figure(figsize = (18, 7))
sns.scatterplot(x = 'College Name', y = 'CGPA',color = "green", data = df, alpha = 0.5)
plt.xticks(rotation = 90)
plt.xlabel('Collegs')
plt.ylabel('CGPA')
plt.title('CGPA variation of Colleges')
plt.show()


```

```python
df['Family Income'].value_counts()

custom_mapping = {'0-2 Lakh': 100000,'2-5 Lakh': 300000,'7 Lakh+': 700000,'5-7 Lakh': 600000}

df['Family Income'] = df['Family Income'].replace(custom_mapping)

cor = df['CGPA'].corr(df['Family Income'])
print("The correlaton between the CGPA and Family Income is :", cor)
```

```python
city = df['City'].value_counts()
city = city.head(10)
plt.figure(figsize=(10, 6))
city.plot(kind='bar', color='skyblue')
plt.xlabel('City')
plt.ylabel('Number of Students')
plt.title('Top 10 Cities with the Highest Number of Students')
plt.xticks(rotation=360, ha='right')
plt.tight_layout()
plt.show()
```

```python
k = df[['CGPA', 'Family Income', 'Experience with python (Months)', 'Expected salary (Lac)']]
sns.pairplot(k)
plt.show()
```

```python
l = df.groupby(['Designation', 'Events']).size().reset_index(name='count')
t = l.pivot(index = 'Designation', columns = 'Events', values = 'count').fillna(0)
plt.figure(figsize = (20,12))
sns.heatmap(t, cmap = 'coolwarm', annot = True, fmt='.0f', linewidths=.5)
plt.title("Event Attendance by feild of Study")
plt.ylabel('Designation')
plt.xlabel('Events')
plt.show()

```

```python
ls = df[df['Leadership- skills'] == 'yes']
ln = df[df['Leadership- skills'] == 'no']

plt.figure(figsize=(8, 6))
plt.hist(ln['CGPA'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('CGPA')
plt.ylabel('Frequency')
plt.title('Distribution of GPA for Students with Leadership Skills (No)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
plt.figure(figsize=(8, 6))
plt.hist(ln['Expected salary (Lac)'], bins=20, color='salmon', edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of GPA for Students with Leadership Skills (No)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
